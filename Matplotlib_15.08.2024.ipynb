## Gunluk Kazanc Bulma (Feature Engeneering)

```python
# Sayısal verilerin seçilmesi
numeric_tips = tips[["total_bill", "day"]]
```

```python
# Günler bazında "total_bill" toplamlarının seçilmesi
day_of_total_bill_sum = numeric_tips.groupby("day")["total_bill"].sum()
```

```python
# Bar Grafiği için günler ve toplamlarının alınması 
days = day_of_total_bill_sum.index
total_bills = day_of_total_bill_sum.values
```

```python
# Bar Grafiğin çizilmesi
fig, ax = plt.subplots(figsize = (6,4))

ax.bar(days, total_bills);
ax.set_title("Total Bill by Day")
ax.set_xlabel("Day")
ax.set_ylabel("Total Bill")

plt.show()
```

## Pie Chart

```python
x = np.array([100, 100, 100, 100])
mylabels = ("Apples", "Bananas", "Cherries", "Melon")

plt.figure(figsize = (8,6))

plt.pie(x, labels = mylabels, labeldistance = 0.5, autopct = "%.2f")
# labeldistance label larin merkese olan uzakligini belirler
# autopct yuzdelik degerle nasil gosterilcegini ayarlar

plt.show()
```

```python
x = np.array([100, 100, 100, 100])
mylabels = ("Apples", "Bananas", "Cherries", "Melon")
myexplode = [0.5, 0, 0, 0]

plt.figure(figsize = (8,6))

plt.pie(x, labels = mylabels, labeldistance = 0.5, autopct = "%.2f", explode = myexplode)
# labeldistance label larin merkese olan uzakligini belirler
# autopct yuzdelik degerle nasil gosterilcegini ayarlar
# explode parcayi ayirir 3 tanesini 0 yaptik yani ayirmadik digerinin mesafesini ayarliyip ayirdik

plt.show()
```

```python
x = np.array([100, 100, 100, 100])
mylabels = ("Apples", "Bananas", "Cherries", "Melon")
myexplode = [0, 0, 0, 0]

plt.figure(figsize = (8,6))

plt.pie(x, labels = mylabels, labeldistance = 0.8, autopct = "%.2f", explode = myexplode, startangle = 45)
# labeldistance label larin merkese olan uzakligini belirler
# autopct yuzdelik degerle nasil gosterilcegini ayarlar
# explode parcayi ayirir 3 tanesini 0 yaptik yani ayirmadik digerinin mesafesini ayarliyip ayirdik
# startangle = 45 45 derece dondurdu

plt.show()
```

```python
plt.figure(figsize=(10,8))
color_list = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode_list = [0.5, 0, 0, 0]

x = np.array([100, 100, 100, 100])
mylabels = ("Apples", "Bananas", "Cherries", "Melon")
myexplode = [0, 0, 0, 0]

plt.figure(figsize = (8,6))

plt.pie(x, 
        labels = mylabels, 
        labeldistance = 0.5, 
        autopct = "%.1f",
        startangle = 45,
        shadow = True,
        pctdistance = 0.3,
        colors = color_list,
        explode = explode_list)

plt.legend(loc = (1,1))

plt.show()
```

## Data Setinden Pie Chart Olusturma

```python
plt.figure(figsize=(10,8))

sex_counts = tips["sex"].value_counts()
color_list = ["lightcoral", "lightskyblue"]
explode_list = [0.1, 0]

plt.pie(sex_counts, 
        labels = sex_counts.index, 
        labeldistance = 0.6, 
        autopct = "%.1f",
        startangle = 90,
        shadow = True,
        pctdistance = 0.5,
        colors = color_list,
        explode = explode_list)

plt.legend(loc = (1,0.7))

plt.show()
```
